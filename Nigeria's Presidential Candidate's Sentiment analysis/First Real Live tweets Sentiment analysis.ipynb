```python
# Import the dependencies required
import sys
import tweepy
import pandas as pd
import textblob
import matplotlib.pyplot as plt
from __future__ import print_function, division
%matplotlib inline
```

```python
# Create the api

    # the authenticaation variables
consumer_key = "xxxxxxxxxxxxxxxxxxx"
consumer_secret = "xxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
access_token = "xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
access_token_secret = "xxxxxxxxxxxxxxxxxxxxxxxxxx"
```

```python
# Authentication object
authenticate = tweepy.OAuthHandler(consumer_key, consumer_secret)
authenticate.set_access_token(access_token, access_token_secret)

# create api
twitter_api = tweepy.API(authenticate)
```

```python
# stream tweets and put it in a dictonary

    # key words to search and the number of tweets to search
search_key_words = str(raw_input("Enter the key word or hashtag to search: "))
search_no = int(raw_input("Enter the number of tweets to search: "))

# tweets
tweets = tweepy.Cursor(twitter_api.search, search_key_words).items(search_no)

# create a dictionary of the tweets and polarity
tweets_dict2 =[]
for tweet in tweets:
    try:
        tweet = tweet.text
        tweet_text = tweet.strip().encode("ascii", "ignore")
        blob = textblob.TextBlob(tweet_text)
        polarity = blob.sentiment.polarity
        tweets_dict2.append((tweet_text, polarity))
    except:
        print("error")
```

```python
# covert the dictionary to a dataframe
tweets_df2 = pd.DataFrame(tweets_dict2)
```

```python
# Rename the columns
tweets_df2.columns = ["tweets", "polarity"]
```

```python
# Now let us analyze sentiments


    # key words to search and the number of tweets to search
search_key_words = str(raw_input("Enter the key word or hashtag to search: "))
search_no = int(raw_input("Enter the number of tweets to search: "))

def percentage(part, whole):
    return((part/whole) * 100)

# polarity
positive = 0
neutral = 0
negative = 0
polarity = 0



# tweets
tweets = tweepy.Cursor(twitter_api.search, search_key_words).items(search_no)


for tweet in tweets:
    #print(tweet.text)
    blob_tweet = textblob.TextBlob(tweet.text)
    
    
    if(blob_tweet.sentiment.polarity > 0.0):
        positive += 1
        
    elif(blob_tweet.sentiment.polarity == 0.0):
        neutral += 1
    
    elif(blob_tweet.sentiment.polarity < 0.0):
        negative += 1

        
# calculate percentage
positive = percentage(positive, search_no)
neutral = percentage(neutral, search_no)
negative = percentage(neutral, search_no)

# round up to 2 decimal places
positive = format(positive, ".2f")
neutral = format(neutral, ".2f")
negative = format(negative, ".2f")

# Draw a pie chart
sizes = [positive, neutral, negative]
labels = ["Positive sentiments " + str(positive) + "%", "Neutral sentiments " + str(neutral) + "%", "Negative sentiments " + str(negative) + "%"]
colors = ["yellowgreen", "pink", "gold"]
patches, text = plt.pie(sizes, colors=colors, startangle=90)
plt.legend(patches, labels, loc="best")
plt.title("Sentiments of tweets concerning " + search_key_words)
plt.show()
```
